```python
import json
from matplotlib import pyplot as plt
import numpy as np
import os
```

```python
%matplotlib inline
def getConfig(filedir, exp):
    dir_name = 'Experiment_Data_%s/%d/'%(filedir, exp)
    filename = dir_name + 'config.json'
    with open(filename) as json_file:
        return json.load(json_file)

def loadData(filedir='WINE'):
    def getAccuracy(filedir):
        filename = filedir + 'metrics.json'
        with open(filename) as json_file:
            data = json.load(json_file)
        acc = data['val_accuracy']['values']
        loss = data['val_loss']['values']
        return acc, loss
        
        
    def getConfig(filedir):
        filename = filedir + 'config.json'
        with open(filename) as json_file:
            return json.load(json_file)
        
    val_acc = []
    val_loss = []
    prec = []
    shots = []
    
    dir_name = 'Experiment_Test_%s'%filedir
    
    dirs = os.listdir(dir_name)
    dirs.remove('_sources')
    for directory in dirs:
        filedir = dir_name + '/' + directory + '/'
        try:
            acc, loss = getAccuracy(filedir)
        except:
            print('Empty entry', filedir)
        else:
            val_acc.append(acc[0:50])
            val_loss.append(loss[0:50])
            config = getConfig(filedir)
            prec.append(config['precision'])
            shots.append(config['shots'])
    val_acc, val_loss, prec, shots = zip(*sorted(zip(val_acc, val_loss, prec, shots)))
    return np.array(val_acc), np.array(val_loss), np.array(prec), np.array(shots)

acc, loss, prec, shots = loadData()
print(np.mean(acc, axis=1).shape, prec.shape)
bits = np.log2(prec)
plt.plot(bits, acc, 'ro')
plt.xlabel('Precision (bits)')
plt.ylabel('Accuracy (%)')
plt.show()

bits = np.log2(prec)
plt.plot(bits, loss, 'ro')
plt.xlabel('Precision (bits)')
plt.ylabel('Loss')
plt.show()
#shots, acc, loss, prec = zip(*sorted(zip(shots, acc, loss, prec)))
plt.plot(shots, np.mean(acc, axis=1))
plt.show()
plt.plot(shots, np.std(acc, axis=1))
```

```python
%matplotlib inline
def getConfig(filedir, exp):
    dir_name = 'Experiment_Data_%s/%d/'%(filedir, exp)
    filename = dir_name + 'config.json'
    with open(filename) as json_file:
        return json.load(json_file)

def loadData(filedir='WINE_NOISY'):
    def getAccuracy(filedir):
        filename = filedir + 'metrics.json'
        with open(filename) as json_file:
            data = json.load(json_file)
        acc = data['val_accuracy']['values']
        loss = data['val_loss']['values']
        return acc, loss
        
        
    def getConfig(filedir):
        filename = filedir + 'config.json'
        with open(filename) as json_file:
            return json.load(json_file)
        
    val_acc = []
    val_loss = []
    prec = []
    shots = []
    
    dir_name = 'Experiment_Test_%s'%filedir
    
    dirs = os.listdir(dir_name)
    dirs.remove('_sources')
    for directory in dirs:
        filedir = dir_name + '/' + directory + '/'
        try:
            acc, loss = getAccuracy(filedir)
        except:
            print('Empty entry', filedir)
        else:
            val_acc.append(acc)
            val_loss.append(loss)
            config = getConfig(filedir)
            prec.append(config['sigma'])
            shots.append(config['shots'])
    val_acc, val_loss, prec, shots = zip(*sorted(zip(val_acc, val_loss, prec, shots)))
    return np.array(val_acc), np.array(val_loss), np.array(prec), np.array(shots)

acc, loss, prec, shots = loadData()
bits = np.log2(2/prec)
plt.plot(bits, np.mean(acc, axis=1), 'r.')
plt.xlabel('Precision (bits)')
plt.ylabel('Accuracy (%)')
plt.title('Effect of Bit Precision On Average Network Accuracy')
plt.savefig('BitPrecNetAcc.jpg', dpi=300)
plt.show()

bits = np.log2(2/prec)
plt.plot(bits, loss, 'r.')
plt.xlabel('Precision (bits)')
plt.ylabel('Loss')
plt.title('Effect of Bit Precision On Network Loss')
plt.savefig('BitPrecNetLoss.jpg', dpi=300)
plt.show()
#shots, acc, loss, prec = zip(*sorted(zip(shots, acc, loss, prec)))
plt.plot(shots, np.mean(acc, axis=1), 'r.')
plt.xlabel('Shots')
plt.ylabel("Average Accuracy")
plt.title("Effect of Shots On Average Network Accuracy")
plt.savefig('ShotsNetAcc.jpg', dpi=300)
plt.show()
plt.plot(shots, np.std(acc, axis=1), 'r.')
plt.xlabel('Shots')
plt.ylabel("$\Delta$Accuracy")
plt.title("Effect of Shots On Network Accuracy Uncertainty")
plt.savefig('ShotsNetAccUnc.jpg', dpi=300)
plt.show()
plt.plot(bits, np.std(acc, axis=1), 'r.')
plt.xlabel('Precision (bits)')
plt.ylabel("$\Delta$Accuracy")
plt.title("Effect of Bits On Network Accuracy Uncertainty")
plt.savefig('BitPrecNetAccUnc.jpg', dpi=300)

```

```python
def getAccuracy(filedir):
        filename = filedir + 'metrics.json'
        with open(filename) as json_file:
            data = json.load(json_file)
        acc = data['val_accuracy']['values']
        acc_tr = data['accuracy']['values']
        return [acc, acc_tr]
def getLoss(filedir):
        filename = filedir + 'metrics.json'
        with open(filename) as json_file:
            data = json.load(json_file)
        acc = data['val_loss']['values']
        acc_tr = data['loss']['values']
        return [acc, acc_tr]

filedir = 'Experiment_Data_WINE/224/'

acc = getAccuracy(filedir)
loss = getLoss(filedir)

plt.plot(acc[0], label='Testing Accuracy');
plt.plot(acc[1], label='Training Accuracy');
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.title("Accuracy Training Results of WINE")
plt.legend()
plt.savefig('Exp224Accuracy.jpg', dpi=300)
plt.show()
plt.plot(loss[0], label='Testing Loss');
plt.plot(loss[1], label='Training Loss')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend()
plt.title("Loss Training Results of WINE")
plt.savefig('Exp224Loss.jpg')
```
